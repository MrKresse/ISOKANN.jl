```julia
using Pkg, Revise
Pkg.activate("..")
using ISOKANN
```

```julia
iso = ISORun(
    sim=MollyLangevin(sys=PDB_ACEMD(), dt=2e-3, T=1e-1, temp=298., gamma=100.), 
    loggers=[ISOKANN.EigenvalueLogger()], 
    opt=ISOKANN.AdamRegularized(2e-4,1e-5), 
    nd=10000)
```

```julia
run!(iso)
```

```julia
plot_learning(iso)
```

```julia
extrema(iso.model(iso.data[1]))
```

```julia
using Plots
plot(reduce(hcat, collect.(iso.loggers[1].values))', title="Eigenvalue estimates", xlabel="n", ylabel="\\lambda", labels=["extremal" "regression"])
```
